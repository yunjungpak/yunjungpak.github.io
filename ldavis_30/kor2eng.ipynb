```python
import json
#google translate
from googletrans import Translator
#timeout error
import time
```

```python
f = open('lda.json', encoding='utf-8')
data = json.load(f)
```

```python
kor_term_list = data['tinfo']['Term']
```

```python
translator = Translator()
kor_eng = {}

for i in range(len(kor_term_list)):
    
    pair = []
    
    if i % 50 == 0:
        time.sleep(3)
        
    kor_term = kor_term_list[i]
    
    try:
        eng_term = translator.translate(kor_term).text
    except AttributeError:
        print('attError')
        eng_term = kor_term
        
    pair.append(kor_term)
    pair.append(eng_term)

    kor_eng[i] = pair
    
```

```python
eng_term_list = []

for i in range(len(kor_eng)):
    eng_term_list.append(kor_eng[i][1])
```

```python
data['tinfo']['Term'] = eng_term_list

with open('lda_eng.json', 'w') as outfile:
    json.dump(data, outfile)
```
